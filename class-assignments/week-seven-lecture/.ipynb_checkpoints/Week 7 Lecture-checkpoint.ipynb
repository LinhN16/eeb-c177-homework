```python
import csv

family = "Labridae"

csv_file = csv.reader(open('fishdata.csv', 'r'), delimiter= ',')
output_file = open('LabridaeData.txt', 'w+')

for column in csv_file:
    if family == column [1]:
        data1= column
        for col in range(4):
            output_file.write(data1[col] + ',')
        output_file.write(data1[4] + '\n')

output_file.close()
```

```python
import numpy as np

lengths =[]

with open('LabridaeData.txt', 'r') as txtfile:
    extraction = csv.reader(txtfile)
    for row in extraction:
        if row[3] != 'NA':
            lengths.append(float(row[3]))

Labridae_meanlength = np.mean(lengths)
print ('The mean length of the Labridae family is', Labridae_meanlength)
```
