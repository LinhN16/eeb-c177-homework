```python
#create a figure object (width, height) in pixels
fig = plt.figure(figsize=(10.0,3.0))

#title that goes above entire panel
fig.suptitle('Panel', va='bottom', fontsize=16)

#create 3 subholder variables (rows, columns, position)
sbplot1 = fig.add_subplot(1, 3, 1)
sbplot2 = fig.add_subplot(1, 3, 2)
sbplot3 = fig.add_subplot(1, 3, 3)

#add in our 3 subplots with y- axis to differentiate
sbplot1.set_ylabel('average')
sbplot1.plot(np.mean(data, axis=0), '--')

sbplot2.set_xlabel('time (days)')
sbplot2.set_ylabel ('max')
sbplot2.plot(np.max(data, axis=0), '-o')

sbplot3.set_ylabel ('min')
sbplot3.plot(np.min(data, axis=0),'--*')

#search matplotlib.plyplot functions to tweak font size / color

#this automatically configures the size, amazing
fig.tight_layout()
plt.show()
```

```python
##get all even numbers from 1 to 100
evens = []
odds = []
for ii in range (1,101):
    if (ii % 2 == 0):
        evens.append(ii)
    elif (ii % 2 == 1):
        odds.append(ii)
        
print ("odd numbers:", odds)
print ('\n')
print ("even numbers:", evens)
```

```python
#how to write a functional for loop

evens = [i for i in range(101) if i % 2 == 0]
print (evens)

print ('\n')

square_nums_10 = [i * i for i in range(20) if i % 2 == 0]
print (square_nums_10)
```
